## Question 2
#### Explain how patches relate to the idea of committing a change to the history of your code?

A patch records the differences between two versions of our code such as what lines were added, removed, or modified. When we commit in Git, we are taking the changes that we have staged and are saving that data.

## Question 3
#### What is the difference between a branch and a fork?

A branch is creates a new line of development inside the same repository but a fork creates a new independent copy of the whole repository.

## Question 4
#### How are diff and patch related? What’s the difference?

Diff is the output after comparing two files; it shows what lines were added, changed, or removed. Diff generates the patch file that has this information. Patch uses this file to apply these changes to a new file without overwriting the original.
